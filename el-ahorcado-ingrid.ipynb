# JUEGO: EL AHORCADO 

**FASE INICIAL** 


OBJETIVO:

- Elegir palabra secreta

- Crear tablero, mostrar la palabra oculta con guiones (_ _ _ _ _)

- Pedir letras al jugador (input)

- Crear lista con letras usadas vacia[]

- El jugador intenta adivinarla letra por letra.

- Si acierta, se muestran esas letras correctas.

- Número de intentos 
 
- Si falla, pierde un intento.

- restar intentos si falla

- El juego termina cuando:

            adivina toda la palabra, se queda sin intentos o decide salir con esc.

# SEGUNDO CODIGO CON RANDOM Y ESTRUCTURA MEJORADA:
PENDIENTE:

- INCLUIR EMOTICONOS Y COLORES

```python
## VARIABLES A TENER EN CUENTA##

import random
palabras = [
    # Frutas (25)
    "manzana", "pera", "platano", "naranja", "fresa",
    "melon", "sandia", "uva", "cereza", "mango",
    "papaya", "kiwi", "limon", "mandarina", "higo",
    "ciruela", "frambuesa", "granada", "coco", "pina",
    "melocoton", "pomelo", "mora", "datil", "guayaba",

    # Ropa (25)
    "camiseta", "pantalon", "vestido", "falda", "chaqueta",
    "abrigo", "bufanda", "guantes", "calcetines", "zapatos",
    "botas", "sandalias", "corbata", "cinturon", "sombrero",
    "gorra", "pijama", "bañador", "jersey", "sudadera",
    "impermeable", "chaleco", "zapatillas", "mocasines", "camisa",

    # Animales (25)
    "perro", "gato", "caballo", "vaca", "cerdo",
    "oveja", "gallina", "conejo", "raton", "serpiente",
    "cocodrilo", "elefante", "jirafa", "leon", "tigre",
    "oso", "lobo", "zorro", "delfin", "ballena",
    "aguila", "loro", "camello", "gorila", "tortuga",

    # Cosas (25)
    "mesa", "silla", "lampara", "telefono", "ordenador",
    "televisor", "nevera", "microondas", "espejo", "reloj",
    "mochila", "paraguas", "tijeras", "linterna", "camara",
    "teclado", "auriculares", "botella", "maleta", "cuaderno",
    "lapicero", "calculadora", "altavoz", "bicicleta", "escalera"
]


palabra_secreta = random.choice(palabras)               # 1. Palabra secreta (import random)
tablero = ["_"] * len(palabra_secreta)                  # 2. Crear el tablero * num de letras
letras_usadas = set()                                      # 3. almacenar las letras usadas por el jugador
intentos = 6 

salir = False

while intentos > 0 and not salir:
                                    ##estado al inicio de cada turno

    print(f"\nPalabra: {' '.join(tablero)}")
    print(f"Letras usadas: {', '.join(sorted(letras_usadas)) if letras_usadas else 'Ninguna'}")
    print(f"Intentos restantes: {intentos}")

    letra = input("Bienvenida al juego El Ahorcado! Escribe una letra para empezar o las letras esc para salir")

    if letra.lower() == "esc":
        print("¡Hasta la próxima!")
        salir = True
        continue
    
    if len(letra) > 1:
        print("¡Error! Debes introducir una sola letra:")
        continue

    if not letra.isalpha():          ##isalpha: bloquea numeros y simbolos,(4!@) devuelve True si el caracter es una letra
        print("Eso no es una letra valida. ¡Introduce solo letras!")
        continue
    
    if letra in letras_usadas:
        print("Esa letra ya la usaste....Escribe otra letra.")
        continue

    if letra in palabra_secreta:
        print(f"¡Bien! La letra '{letra}' está en la palabra")
        letras_usadas.add(letra)
        for i in range(len(palabra_secreta)):
            if palabra_secreta[i] == letra:
                tablero[i] = letra                            ##letras acertadas hasta aqui
        print(f"Letras acertadas: {', '.join(l for l in letras_usadas if l in palabra_secreta)}")

    if letra not in palabra_secreta:
        letras_usadas.add(letra)                             ## seañade tmb para sumar letras fallidas
        intentos -= 1 
        print(f"La letra '{letra}' no está en la palabra")   ##letras falladas hasta aqui
        print(f" letras falladas: {', '.join(l for l in letras_usadas if l not in palabra_secreta)}")
        print(f"Te quedan {intentos} intentos")

    if "_" not in tablero:
        print("¡Bien hecho, has ganado la partida!!")
        break
if intentos == 0:
    print("Lo siento has consumido todos tus intentos..¡has perdido!")  
```
