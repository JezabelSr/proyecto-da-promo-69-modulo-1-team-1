```python
# IMPORT RANDOM

import random 

# DEFINICIÓN DE VARIABLES 

lista = ["piedra","papel","tijera"]
puntos_PC = 0 
puntos_jugador = 0
puntos_necesarios = 3
ronda = 0
salir = False

# BUCLE PARA QUE SE REPITA HASTA QUE UNO LLEGUE A 3

while True:

    print("¡Bienvenida al juego de Piedra, Papel o Tijera! 🎮🏆🎉")
    print("Quien llegue primero a 3 puntos ganará ¡Buena suerte! 🍀💚\n")
    print("\033[3mSi en cualquier momento quieres salir, escribe 'salir'\n\033[0m")

    while puntos_jugador < puntos_necesarios and puntos_PC < puntos_necesarios:

        #ELECCION PC 

        eleccion_PC = random.choice(lista)

        # ELECCION JUGADOR

        while True:
            eleccion_jugador = input("¿Piedra, papel o tijera? 🪨📄✂️").lower()
            if eleccion_jugador in lista: 
                ronda +=1
                print("\033[35m---------------------------------------------\033[0m")
                print(f"\033[35m------------------ Ronda {ronda} ------------------\n\033[0m")
                print(f"Has elegido {eleccion_jugador}. El PC ha elegido {eleccion_PC}.")
                break 

            elif eleccion_jugador == "salir": 
                print("Muchas gracias por jugar 🪨📄✂️ ¡nos vemos pronto! 👋😊")
                salir = True
                break

            else:
                print ("Respuesta no válida, por favor vuelve a intentarlo")

        if salir:
            break

        # DETERMINACIÓN DEL GANADOR 

        if eleccion_jugador == eleccion_PC:
            print("🤝 ¡Empate, prueba otra vez! ✨")
        elif ((eleccion_jugador == "piedra" and eleccion_PC == "tijera") or 
        (eleccion_jugador == "papel" and eleccion_PC == "piedra") or 
        (eleccion_jugador == "tijera" and eleccion_PC == "papel")):
            print ("🎊 ¡Enhorabuena, ganaste la ronda! 🔥")
            puntos_jugador +=1
        else: 
            print("💔 Vaya, has perdido, ¡inténtalo de nuevo! 💪")
            puntos_PC +=1
        print(f"Marcador: Tú {puntos_jugador * '⭐' if puntos_jugador != 0 else '0'} vs PC {puntos_PC * '⭐' if puntos_PC != 0 else '0'}\n")

    if salir:
        break

    # FIN DEL JUEGO 

    if puntos_jugador == puntos_necesarios:
        print("\033[30;47m¡Enhorabuena, ganaste el juego!🏆🎊🥳\033[0m\n")
    else: 
        print("\033[30;47mVaya, el PC ganó en esta ocasión 🤖 ¡otra vez será! 🌈\033[0m\n")

    respuesta = input("¿Quieres jugar de nuevo? Pulsa 'S' para sí o 'N' para no").upper()
    while respuesta != "N" and respuesta != "S":
        print ("Respuesta no válida, por favor pulsa 'S' para sí o 'N' para no")
        respuesta = input ("¿Quieres jugar de nuevo? Pulsa 'S' para sí o 'N' para no").upper()

    if respuesta == "N":
        print(f"Muchas gracias por jugar 🪨📄✂️ ¡nos vemos pronto! 👋😊")
        break
    else: 
        puntos_PC = 0 
        puntos_jugador = 0
        ronda = 0
        salir = false 
```
