```python
palabras = ["gato", "perro", "casa", "arbol", "libro", "mesa", "silla", "puerta", "ventana", "jardin",
    "playa", "monte", "rio", "lago", "mar", "cielo", "nube", "sol", "luna", "estrella",
    "flor", "hoja", "rama", "raiz", "fruto", "manzana", "pera", "uva", "limon", "naranja",
    "coche", "tren", "avion", "barco", "bici", "moto", "camion", "autobus", "metro", "taxi",
    "pan", "leche", "queso", "huevo", "arroz", "pasta", "sopa", "carne", "pollo", "pescado",
    "rojo", "azul", "verde", "negro", "blanco", "amarillo", "naranja", "morado", "rosa", "gris",
    "perla", "coral", "piedra", "arena", "tierra", "fuego", "agua", "viento", "hielo", "vapor",
    "tigre", "leon", "oso", "lobo", "zorro", "aguila", "serpiente", "delfin", "ballena", "pulpo",
    "cama", "sofa", "espejo", "reloj", "lampara", "cuadro", "alfombra", "cortina", "armario", "cajón",
    "musica", "baile", "pintura", "teatro", "cine", "deporte", "futbol", "tenis", "natacion", "ciclismo"]


import os
import random

palabra = random.choice(palabras)
letras_probadas = set()
intentos = 6

dibujos = ["""-----""" ,
"""----- 
|            |""",
"""----- 
|            | 
|            0""",
"""----- 
|            | 
|            0
             | """,
"""----- 
|            | 
|            0
|          / | \\""",
"""----- 
|            | 
|            0
|          / | \\
|           / \\""",
"""----- 
|            | 
|            0
|          / | \\
|           / \\
|"""]




print("\033[1;36m 🎮 ═══════════════════════════════ 🎮 \033[0m")
print("\033[1;36m       BIENVENIDO AL AHORCADO          \033[0m")
print("\033[1;36m 🎮 ═══════════════════════════════ 🎮 \033[0m")

salir = False
while True:
    palabra = random.choice(palabras)
    letras_probadas = set()
    intentos = 6
    contador_pistas = 0

    nombre = input ("Bienvenido al Ahorcado. ¿Cómo te llamas?")
    print(f"\033[1;30;43m {nombre.capitalize()}, puedes pulsar '1' en cualquier momento para salir del juego \033[0m")
    print(f"\033[1;30;44m Si te atascas puedes pedir hasta dos pistas en toda la partida pulsando el '2' \033[0m")
    while intentos > 0 and not all([i in letras_probadas for i in palabra]):
        print(f"La palabra tiene {len(palabra)} letras")
        print(" ".join([i if i in letras_probadas else "_" for i in palabra]))
        print(f"\033[35mVidas restantes: {'❤️ ' * intentos}\033[0m")
        print("Has usado ya estas letras: " + " ".join(["\033[32m" + i + "\033[0m" if i in palabra else "\033[31m" + i + "\033[0m" for i in letras_probadas]))
        letra = input("Introduce una letra: ")

        if letra == "1":
            print("¡Hasta pronto!")
            salir = True
            break
        elif letra == "2":
            if contador_pistas <2:
                no_adivinadas = [i for i in palabra if i not in letras_probadas]
                pista = random.choice (no_adivinadas)
                print(f"\033[1;4;30;44m La palabra contiene la letra '{pista}' \033[0m")
                contador_pistas +=1
            else: 
                print("\033[1;4;30;44m Lo siento, has agotado ya tus 2 pistas \033[0m")
        elif not letra.isalpha():
            print("\033[34m Por favor introduce una letra válida (abcdefghijklmnñopqrstuvwxyz) \033[0m")

        elif letra in letras_probadas:
            print("\033[33m ¡Esta letra ya la habías probado, prueba otra!\033[0m")
        elif letra in palabra:
            letras_probadas.add(letra)
            print(f"\033[32m ¡Muy bien, la {letra} si está!\033[0m")
        else:
            print(f"\033[31m ¡Vaya! La {letra} no está , prueba otra letra\033[0m")
            intentos -=1
            letras_probadas.add(letra)
        print(dibujos[6 - intentos])

    if salir: break

    if all([i in letras_probadas for i in palabra]):
        print(f"\033[1;30;42m 🏆¡Enhorabuena {nombre.capitalize()}!¡Has ganado!🏆 Si quieres volver a jugar pulsa 's', si no pulsa 'n' \033[0m")
    else:
        print(f"\033[1;30;41m 💀¡Vaya {nombre.capitalize()}! Has perdido...💀 La palabra era: {palabra}. Pero puedes volver a intentarlo pulsando 's', si no pulsa 'n' \033[0m")

    respuesta = input("¿Quieres jugar de nuevo? (s/n): ")
    while respuesta != "n" and respuesta != "s":
        print ("Respuesta no válida, por favor introduce 's' para seguir jugando, o 'n' para dejar el juego")
        respuesta = input("¿Quieres jugar de nuevo? (s/n): ")

    if respuesta == "n":
        print(f"\033[1;30;42m 🎉¡GRACIAS POR JUGAR AL AHORCADO {nombre.upper()}!.¡HASTA PRONTO!🎉 \033[0m")
        break
```
