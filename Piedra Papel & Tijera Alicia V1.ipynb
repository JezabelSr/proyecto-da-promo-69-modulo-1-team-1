```python
# IMPORT RANDOM

import random 

# DEFINICIÓN DE VARIABLES 

lista = ["piedra","papel","tijera"]
puntos_PC = 0 
puntos_jugador = 0
puntos_necesarios = 3
ronda = 0

# BUCLE PARA QUE SE REPITA HASTA QUE UNO LLEGUE A 3

print("¡Bienvenida al juego de Piedra, Papel o Tijera! 🎮🏆🎉")
print("El primero en llegar a 3 puntos gana ¡Buena suerte! 🍀💚\n")

while puntos_jugador < puntos_necesarios and puntos_PC < puntos_necesarios:

    #ELECCION PC 

    eleccion_PC = random.choice(lista)

    # ELECCION JUGADOR

    while True:
        eleccion_jugador = input("¿Piedra, papel o tijera? 🪨 📄 ✂️").lower()
        if eleccion_jugador in lista: 
            ronda +=1
            print(f"\n----------- Ronda {ronda} -----------")
            print(f"Has elegido {eleccion_jugador}. El PC eligió {eleccion_PC}.")
            break
        else:
            print ("Respuesta no válida, por favor vuelve a intentarlo")

    # DETERMINACIÓN DEL GANADOR 

    if eleccion_jugador == eleccion_PC:
        print("🤝 ¡Empate, prueba otra vez! ✨")
    elif ((eleccion_jugador == "piedra" and eleccion_PC == "tijera") or 
    (eleccion_jugador == "papel" and eleccion_PC == "piedra") or 
    (eleccion_jugador == "tijera" and eleccion_PC == "papel")): 
        print ("🎊 ¡Enhorabuena, ganaste la ronda! 🔥")
        puntos_jugador +=1
    else: 
        print("💔 Vaya, esta vez perdiste, ¡inténtalo de nuevo! 💪")
        puntos_PC +=1
    print(f"Marcador: Tú {puntos_jugador * '⭐' if puntos_jugador != 0 else '0'} vs PC {puntos_PC * '⭐' if puntos_PC != 0 else '0'}\n")

# FIN DEL JUEGO 

if puntos_jugador == puntos_necesarios:
    print("¡Enhorabuena, ganaste el juego!🏆🎊🥳")
else: 
    print("Vaya, el PC ganó en esta ocasión 🤖 ¡otra vez será! ✨")
```
